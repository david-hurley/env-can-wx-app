```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
df = pd.read_csv('station-metadata-processed.csv')
```

```python
from folium.plugins import FastMarkerCluster
#grab a random sample from df
map = folium.Map(location=(57,-106),zoom_start=3,tiles='Stamen Terrain')
callback = ('function (row) {' 
                'var circle = L.circle(new L.LatLng(row[0], row[1]), {color: "red",  radius: 20000});'
                'return circle};')
map.add_child(FastMarkerCluster(df[['Latitude','Longitude']].values.tolist(), callback=None))
map
```

```python
map.save('wx_location.html')
```

```python
locationlist = df[['Latitude','Longitude']].values.tolist()
map = folium.Map(location=(57,-106),zoom_start=3,tiles='Stamen Terrain')
    
mc = MarkerCluster()

for row in df.itertuples():
    mc.add_child(folium.Marker(location=[row.Latitude,  row.Longitude],
                 popup=row.Name))
 
map.add_child(mc)
map.save('station_locations.html')
```
